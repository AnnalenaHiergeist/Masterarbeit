### Detect Language

```python
from langdetect import detect
```

```python
def detect_language(text):
    try:
        language = detect(text)
        return language
    except:
        return "Unbekannt"  # Wenn die Sprache nicht erkannt wird

# Die Spalte "Sprache" erstellen, um die erkannte Sprache zu speichern
df['language_tweets'] = df['tweet_text'].apply(detect_language)
df['language_retweets'] = df['retweet_text_retweeted_tweet'].apply(detect_language)
df['language_zitate'] = df['zitat_text'].apply(detect_language)
```

### Preprocessing

Zun√§chst nur wenige Preprocessing-Schritte, um Annotation durchf√ºhren zu k√∂nnen
-> weitere Schritte folgen dann vor der Durchf√ºhrung der jeweiligen Analysen

```python
import pandas as pd
df = pd.read_excel('df_de_raw.xlsx')
```

```python
import html
import re
```

```python
def cleanText(text):
    # HTML Escape-Zeichenfolgen entschl√ºsseln wie &amp
    text = html.unescape(text)

    # Retweet-Indikatoren (RT) und zugeh√∂rige Leerzeichen entfernen
    text = re.sub(r'RT[\s]+', '', text)

    # URLs entfernen
    text = re.sub('http[s]?://(?:[a-zA-Z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-fA-F][0-9a-fA-F]))+', '', text, flags=re.MULTILINE)

    # Zeilen mit numerischen Mustern am Anfang entfernen
    text = re.sub(r'\n[0-9].[0-9]', '', text)
    text = re.sub(r'\n[0-9][0-9]', '', text)
    text = re.sub(r'\n[0-9]', '', text)
    text = re.sub(r'\n[0-9].[0-9][0-9][0-9]', '', text)

    # Einzelne Zeilenumbr√ºche entfernen
    text = re.sub(r'\n', ' ', text)

    # Mehrere aufeinander folgende Zeilenumbr√ºche durch ein Leerzeichen ersetzen
    text = re.sub(r'\n+', ' ', text)
    
    # Mehrfache Leerzeichen durch ein Leerzeichen ersetzen
    text = re.sub(r'\s+', ' ', text)

    # Zwei oder mehr aufeinander folgende Punkte durch ein Leerzeichen ersetzen
    text = re.sub(r'\.{2,}', ' ', text)

    # "&" durch das Wort "und" ersetzen
    text = text.replace("&", "und")

    return text
```

```python
df['tweet_text'] = df['tweet_text'].astype(str)
df['retweet_text_retweeted_tweet'] = df['retweet_text_retweeted_tweet'].astype(str)
df['zitat_text'] = df['zitat_text'].astype(str)
df['tweet_creator_bio'] = df['tweet_creator_bio'].astype(str)
df['retweet_bio_creator_retweeted_tweet'] = df['retweet_bio_creator_retweeted_tweet'].astype(str)
df['zitat_bio_creator_qouted_tweet'] = df['zitat_bio_creator_qouted_tweet'].astype(str)
```

```python
df['tweet_text'] = df['tweet_text'].apply(cleanText)
df['retweet_text_retweeted_tweet'] = df['retweet_text_retweeted_tweet'].apply(cleanText)
df['zitat_text'] = df['zitat_text'].apply(cleanText)
df['tweet_creator_bio'] = df['tweet_creator_bio'].apply(cleanText)
df['retweet_bio_creator_retweeted_tweet'] = df['retweet_bio_creator_retweeted_tweet'].apply(cleanText)
df['zitat_bio_creator_qouted_tweet'] = df['zitat_bio_creator_qouted_tweet'].apply(cleanText)

df['tweet_creator_bio'] = df['tweet_creator_bio'].str.replace(r'^-900;;', '', regex=True)
df['retweet_bio_creator_retweeted_tweet'] = df['retweet_bio_creator_retweeted_tweet'].str.replace(r'^-900;;', '', regex=True)
df['zitat_bio_creator_qouted_tweet'] = df['zitat_bio_creator_qouted_tweet'].str.replace(r'^-900;;', '', regex=True)
```

Hashtags extrahieren:

```python
df["hashtags_tweet"] = df["tweet_text"].apply(lambda x: re.findall(r"#(\w+)", x))
df["hashtags_retweet"] = df["retweet_text_retweeted_tweet"].apply(lambda x: re.findall(r"#(\w+)", x))
df["hashtags_zitat"] = df["zitat_text"].apply(lambda x: re.findall(r"#(\w+)", x))
```
